```python
# Instalar las librerías necesarias
!pip install PyPDF2 transformers torch

# Importar las librerías necesarias
from PyPDF2 import PdfReader
from transformers import AutoTokenizer, AutoModel
import torch
import pickle

# Definir el modelo y el tokenizador de Hugging Face
model_name = "distilbert-base-uncased"  # Puedes usar otros modelos como BERT
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Función para extraer texto de un archivo PDF
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Función para generar embeddings del texto de un documento
def generate_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():  # No es necesario calcular gradientes en inferencia
        outputs = model(**inputs)
    embeddings = torch.mean(outputs.last_hidden_state, dim=1).squeeze()  # Vector promedio
    return embeddings

# Documentos PDF cargados previamente en Google Colab
pdf_files = ["document1.pdf", "document3.pdf"]  # Reemplaza con los nombres reales si son diferentes

# Generar embeddings para los documentos PDF cargados
document_embeddings = {}
for pdf_file in pdf_files:
    # Extraer texto del PDF
    text = extract_text_from_pdf(pdf_file)
    # Generar embeddings del texto extraído
    embeddings = generate_embeddings(text)
    # Guardar los embeddings en el diccionario
    document_embeddings[pdf_file] = embeddings

print("Embeddings generados para los documentos PDF.")

# Guardar los embeddings en un archivo
with open("pdf_document_embeddings.pkl", "wb") as f:
    pickle.dump(document_embeddings, f)
print("Embeddings guardados en 'pdf_document_embeddings.pkl'.")
```

```python
# Instalar las librerías necesarias
!pip install PyPDF2 nltk

# Importar librerías
from PyPDF2 import PdfReader
import nltk

# Descargar datos necesarios para nltk
nltk.download("punkt")

# Función para extraer texto de un archivo PDF
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Normalizar texto (convertir a minúsculas, eliminar caracteres especiales)
def normalize_text(text):
    text = text.lower()  # Convertir a minúsculas
    text = "".join(char for char in text if char.isalnum() or char.isspace() or char == '.')
    return text

# Lista de archivos PDF
pdf_files = ["documentotexto.pdf"]

# Extraer y normalizar texto de cada archivo PDF
normalized_texts = {}
for pdf_file in pdf_files:
    text = extract_text_from_pdf(pdf_file)
    normalized_text = normalize_text(text)
    normalized_texts[pdf_file] = normalized_text

print("Textos normalizados:", normalized_texts)
```

```python
from nltk.tokenize import sent_tokenize
pdf_file= ["documentotexto.pdf"]

# Dividir el texto en fragmentos (chunking) basados en oraciones
def chunk_text(text, max_chunk_length=100):
    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = []

    for sentence in sentences:
        if len(" ".join(current_chunk)) + len(sentence) <= max_chunk_length:
            current_chunk.append(sentence)
        else:
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentence]
    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Crear chunks para cada texto normalizado
chunked_texts = {}
for pdf_file, text in normalized_texts.items():
    chunks = chunk_text(text, max_chunk_length=100)
    chunked_texts[pdf_file] = chunks

print("Chunks creados:", chunked_texts)
```

```python
# Vectorización de los chunks utilizando el modelo de Hugging Face
def vectorize_chunks(chunks):
    chunk_embeddings = []
    for chunk in chunks:
        inputs = tokenizer(chunk, return_tensors="pt", padding=True, truncation=True)
        with torch.no_grad():
            outputs = model(**inputs)
        embedding = torch.mean(outputs.last_hidden_state, dim=1).squeeze()  # Vector promedio para el chunk
        chunk_embeddings.append(embedding)
    return chunk_embeddings

# Generar vectors para cada conjunto de chunks
all_vectors = {}
for pdf_file, chunks in chunked_texts.items():
    vectors = vectorize_chunks(chunks)
    all_vectors[pdf_file] = vectors

print("Embeddings generados para cada chunk de cada documento:", all_vectors)
```

```python
# Instalar las librerías necesarias
!apt-get install -y ffmpeg
!pip install moviepy transformers torch torchvision

# Importar las librerías necesarias
import moviepy.editor as mp
from transformers import AutoTokenizer, AutoModel
import torch
from torchvision import models, transforms
from google.colab import files
from PIL import Image
import numpy as np

# Función para extraer el audio de un video y guardarlo como archivo .wav
def extract_audio_from_video(video_path, output_audio_path="audio.wav"):
    video = mp.VideoFileClip(video_path)
    video.audio.write_audiofile(output_audio_path)
    return output_audio_path

# Función para extraer frames del video
def extract_frames_from_video(video_path, frame_rate=1):
    video = mp.VideoFileClip(video_path)
    frames = []
    for t in np.arange(0, video.duration, 1.0 / frame_rate):
        frame = video.get_frame(t)
        frame_image = Image.fromarray(frame)
        frames.append(frame_image)
    return frames

# Función para generar embeddings visuales de un frame usando un modelo preentrenado (ResNet)
def generate_image_embeddings(image):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    image = transform(image).unsqueeze(0)  # Añadir dimensión de batch

    # Cargar modelo preentrenado ResNet
    model = models.resnet50(pretrained=True)
    model.eval()

    # Obtener embeddings de la imagen
    with torch.no_grad():
        embeddings = model(image)
    return embeddings

# Cargar el archivo de video en Google Colab
print("Por favor, sube un archivo de video .mp4")
uploaded = files.upload()

# Suponiendo que solo se suba un archivo de video
video_path = next(iter(uploaded.keys()))  # Obtener el nombre del archivo de video

# Extraer el audio del video (opcional)
audio_path = extract_audio_from_video(video_path)
print(f"Audio extraído: {audio_path}")

# Extraer frames del video a 1 frame por segundo (puedes ajustar frame_rate para más o menos frames)
frames = extract_frames_from_video(video_path, frame_rate=1)

# Generar embeddings para cada frame y almacenarlos en una lista
frame_embeddings = [generate_image_embeddings(frame) for frame in frames]

# Mostrar la cantidad de embeddings generados y la forma del primer embedding
print(f"Total de embeddings generados para frames: {len(frame_embeddings)}")
print(f"Forma del primer embedding: {frame_embeddings[0].shape}")
```
